# Dimensionality Reduction System 

Problem Statement:
High-dimensional datasets are difficult to analyze and visualize, often leading to inefficiencies in model performance. Imagine a dataset with hundreds of features where many are redundant. This increases complexity and slows down processing.
To solve this, the team builds an AI-based Dimensionality Reduction System.
The system should:
● Reduce dimensions
● Preserve variance
● Visualize patterns Tool Restrictions
● PCA mandatory

```python
#Import Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

# Dataset
https://www.kaggle.com/datasets/uciml/iris

```python
# Load dataset
iris = load_iris()

# Convert to DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df["Species"] = iris.target

# Convert numeric labels to names
df["Species"] = df["Species"].map({
    0: "setosa",
    1: "versicolor",
    2: "virginica"
})
df
```

```python
#Separate Features & Target
X = df.drop("Species", axis=1)
y = df["Species"]
```

```python
#Standardize the Data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Apply PCA (Core Requirement)
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

# Expected Output
● Reduced dataset
● Visualization 


```python
#Create Reduced Dataset
import pandas as pd

pca_df = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
pca_df["Species"] = y

print(pca_df.head())
```

```python
#Check Variance Preservation
print("Explained Variance Ratio:", pca.explained_variance_ratio_)
print("Total Variance Preserved:", sum(pca.explained_variance_ratio_))
```

```python
#Visualization (Required Output)
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))

sns.scatterplot(
    x="PC1",
    y="PC2",
    hue="Species",
    data=pca_df,
    palette="Set1",
    s=100
)

plt.title("PCA Visualization - Iris Dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.grid(True)
plt.show()
```

```python
#Scree Plot
plt.plot(pca.explained_variance_ratio_, marker='o')
plt.title("Scree Plot")
plt.xlabel("Principal Components")
plt.ylabel("Variance Explained")
plt.show()
```

# Conclusion
The Dimensionality Reduction System successfully addresses the challenges of high-dimensional data by applying PCA to transform the dataset into a lower-dimensional space while retaining most of the important information. By reducing redundant features, the system improves computational efficiency and simplifies analysis without significant loss of variance. The resulting reduced dataset enables clear visualization of patterns and relationships, making it easier to interpret underlying structures. Overall, this approach enhances model performance, speeds up processing, and provides meaningful insights from complex data.
